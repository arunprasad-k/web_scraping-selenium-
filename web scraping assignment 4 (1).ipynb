```python
#removing unwanted text in date list
Date_cric=list(date)
Date=[]
for i in date:
    Date.append(i.replace("\n",","))
```

```python
#checking for length of all list
len(match_title),len(place),len(series),len(Date),len(time)
```

```python
#dataframing using pandas
bcci=pd.DataFrame({})
bcci["match_title"]=match_title
bcci["place"]=place
bcci["series"]=series
bcci["Date"]=Date
bcci["time"]=time
bcci
```
