```python
#connecting to web driver
driver=webdriver.Chrome(r"c://chromedriver.exe")
```

```python
#connecting to web page
driver.get("https://www.digit.in/")
```

```python
#clicking_button
search_button=driver.find_element_by_xpath("//a[@href='https://www.digit.in/top-products/']")
search_button.click()
```

```python
#clicking the top 10 gaming tab
button=driver.find_element_by_xpath("/html/body/div[3]/div/div/div[3]/div[2]/ul/li[26]/a")
button.click()
```

```python
#making empty list for reqired info
product_name=[]
market_status=[]
memory=[]
release=[]
official_website=[]
os=[]
display=[]
processor=[]
price=[]
```

```python
url=driver.find_elements_by_xpath("//div[@class='right-container']/div/a")
href_url=[]
for i in url:
    href_url.append(i.get_attribute("href"))
for url in href_url:
    driver.get(url)
    for k in driver.find_elements_by_xpath("//div[@class='heading-wraper']/h1"):
        product_name.append(k.text)
    #for scraping market_status   
    for l in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[1]/div[1]/b"):
        market_status.append(l.text) 
    #for scraping availabilty   
    for m in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[1]/div[2]/b"):
        release.append(m.text)
    #for scraping official_website    
    for n in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[1]/div[3]/b"):
        official_website.append(n.text)
    #for scraping os    
    try:                  #exception handling for nosuchelementexception
        for o in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[1]/div"):
            os.append(o.text)  
    except NoSuchElementException   as f:
        os.append("NA")
    #for scraping display   
    try:                  #exception handling for nosuchelementexception
        for p in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[2]/div"):
            display.append(p.text)     
    except NoSuchElementException   as e:
        display.append("NA")
    #for scraping processor 
    try:                  #exception handling for nosuchelementexception
        for q in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[3]/div"):
            processor.append(q.text)  
    except NoSuchElementException   as e:
        processor.append("NA")
    #for scrapying memory    
    try:                  #exception handling for nosuchelementexception
        for r in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[3]/div/ul/li[4]/div"):
            memory.append(r.text) #appending the memory
    except NoSuchElementException   as e:
        memory.append("NA")   
    #for scraping price    
    try:                  #exception handling for nosuchelementexception
        for s in driver.find_elements_by_xpath("/html/body/div[7]/div/div/div[6]/div/div[2]/div[4]/div[1]/b"):
            price.append(s.text) 
    except NoSuchElementException   as e:
        price.append("NA")    
   
```

```python
#dataframing using pandas
gaming_laptop=pd.DataFrame({})
gaming_laptop["product_name"]=product_name
gaming_laptop["market_status"]=market_status
gaming_laptop["memory"]=memory
gaming_laptop["release date"]=release
gaming_laptop["official_website"]=official_website
gaming_laptop["os"]=os
gaming_laptop["display"]=display
gaming_laptop["processor"]=processor
gaming_laptop["price"]=price
gaming_laptop
```
